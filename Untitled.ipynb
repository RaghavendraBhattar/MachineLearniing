```python
x = [2,4,6,8,10]
y = [2,4,6,8,10]
x2 = np.arange(4,4)
plt.plot(x2**2)
plt.plot(x,y, 'r.--',label='grphs')
plt.title("python plots", fontdict={'family': 'serif', 'fontsize':20})
plt.xlabel('X axis')
plt.ylabel('y axis')
plt.xticks([2,3,4,5,6,7,8,9,10])
plt.yticks([2,4,6,8,10])
plt.legend()
plt.show()
```
