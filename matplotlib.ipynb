## matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
## functional method

plt.plot(x,y,'r--')
plt.xlabel('X label')
plt.ylabel('y label')
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r')

plt.subplot(1,3,2)
plt.plot(x,y,'g')

plt.subplot(1,3,3)
plt.plot(x,y,'b')
```

```python
# oo

fig = plt.figure()

axes = fig.add_axes([1, 1, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.4, 0.2, 0.4, 0.3])
axes1.set_title('plot1')
axes2.set_title('plot2')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x,y)
    
plt.tight_layout()
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(6,3))

ax = fig.add_axes([0.5,0.5,1,1])
ax.plot(x,y)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(6,3))

ax[0].plot(x,y)
ax[1].plot(x,y)
plt.tight_layout()
```
